## ให้ สร้าง list ของ เวลา ```['12:30:15','13:41:07',....]``` แล้ววนลูปเรียกฟังก์ชั่นคำนวณเวลาเป็น วินาที

**commit ว่า HW3**

**สร้างฟังก์ชั่น**

````python
def to_seconds(time):
  h, m, s =time.split(':')
  total = int(h)*3600 + int(m)*60 + int(s)
  return total
````

**สร้าง list ของเวลา**

````python
times = [
    '12:30:15',
    '13:41:07',
    '12:53:15',
    '00:59:25',
    '11:11:11',
    '16:06:09'
]
````

**วนลูป**

````python
for time in times:
  to_seconds(time)
  print(f'{time} = {total:,} วินาที')
````
